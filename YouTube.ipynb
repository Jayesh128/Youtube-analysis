```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
train=pd.read_csv("data/INvideos.csv")
train.head()
```

```python

data=train.copy()

data=data[data['likes']>900]
plt.figure(figsize=(20,10))
plt.title("views on average youtube video",fontsize=19)
sns.distplot(data['views'])
plt.show()
```

```python
sns.jointplot(x="category_id",y="views",data=train,height=10,ratio=3,color='blue')
plt.show()
```

```python
sns.jointplot(x="likes",y="views",data=train,height=10,ratio=3,color='blue')
plt.show()
```

```python
sns.jointplot(x="comment_count",y="likes",data=train,height=10,ratio=3,color='blue')
plt.show()
```

```python
data=train.copy()

data=data[data['dislikes']>900]
plt.figure(figsize=(15,10))
plt.title("views on average youtube video",fontsize=19)
sns.distplot(data['views'])
plt.show()
```
